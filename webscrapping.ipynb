```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "http://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
#type(soup)
print(soup)
```

```python
title = soup.title.text
print(title)
```

```python
# Print out the text
text = soup.get_text(strip=False)
print(text)
#print(text)
```

```python
soup.find_all('a')
```

```python
#<a href="url">Click</a>
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
# Print the first 10 rows for sanity check
rows = soup.find_all('tr')
print(rows[:5])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
#type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "html.parser").get_text()
print(cleantext)
```

```python
import re
list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('.</?\w*>')
    clean2 = (re.sub(clean, '',str_cells))
    #print(clean2)
    
    list_rows.append(clean2)
#print(clean2)
print(list_rows)
#type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df=df.replace('\\n','',regex=True)
df.head(10)
```

```python
time_list = df6[' Chip Time'].tolist()
#print(time_list)

# You can use a for loop to convert 'Chip Time' to minutes

time_mins = []
s='0'
for i in time_list:
    h = i.split(':')
    math = (int(h[0]) * 3600 + int(h[1]) * 60 + int(s))//60
    time_mins.append(math)
print(time_mins)
```

```python
df6['Runner_mins'] = time_mins
df6.head()
```
